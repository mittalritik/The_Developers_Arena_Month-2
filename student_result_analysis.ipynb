```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students_result.csv")
df.head()
```

# 1. Pie chart for gender distribution

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title("Gender Distribution")
plt.show()
```

# 2. Bar graph of value count of ParentEduc

```python
parent_edu_counts = df['ParentEduc'].value_counts()

plt.figure(figsize=(10, 6))
parent_edu_counts.plot(kind='bar', color='skyblue')
plt.title("Parental Education Distribution")
plt.xlabel("Parental Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Donut Graph of value count of LunchType

```python
lunch_counts = df['LunchType'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightgreen'], wedgeprops={'width': 0.4})
plt.title("Lunch Type Distribution (Donut Chart)")
plt.show()
```

# 4. Pie Chart - Distribution of Test Preparation (TestPrep)

```python
prep_counts = df['TestPrep'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(prep_counts, labels=prep_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'orange'])
plt.title("Test Preparation Course Distribution")
plt.show()
```

# 5. Gender-wise Average Exam Score

```python
df['TotalScore'] = df['MathScore'] + df['ReadingScore'] + df['WritingScore']

avg_score_by_gender = df.groupby('Gender')['TotalScore'].mean()

# Step 3: Pie Chart plot karo
plt.figure(figsize=(6,6))
plt.pie(avg_score_by_gender, labels=avg_score_by_gender.index, autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.title("Average Total Score by Gender")
plt.show()
```

# 6. Bar Graph – Value Count of Parental Marital Status (ParentMaritalStatus)

```python
marital_counts = df['ParentMaritalStatus'].value_counts()

plt.figure(figsize=(6, 5))
marital_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Parental Marital Status Distribution")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 7. Grouped Bar Chart: Average Subject Scores by Weekly Study Hours

```python
avg_scores = df.groupby('WklyStudyHours')[['MathScore', 'ReadingScore', 'WritingScore']].mean()

# Create positions for grouped bars
x = np.arange(len(avg_scores))  # WklyStudyHours positions
width = 0.25  # width of each bar

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - width, avg_scores['MathScore'], width, label='Math', color='skyblue')
plt.bar(x, avg_scores['ReadingScore'], width, label='Reading', color='orange')
plt.bar(x + width, avg_scores['WritingScore'], width, label='Writing', color='lightgreen')

# Labels & Titles
plt.xlabel("Weekly Study Hours")
plt.ylabel("Average Score")
plt.title("Average Subject Scores by Weekly Study Hours")
plt.xticks(x, avg_scores.index)
plt.legend()
plt.show()
```

# Conclusion
## 1. Gender Distribution:
### The dataset indicates that 49.7% of the students are male, while 50.3% are female. This suggests that, in this sample, there are slightly more female students than male students.
## 2. Parental Education Level:
### A significant portion of the students come from families where parents have completed their education up to high school or college. However, only a small percentage of students have parents who have completed graduation or post-graduation.
## 3. Parental Financial Status:
### The majority of students (65%) come from families who can afford a standard lunch, indicating that most students are financially stable. On the other hand, a considerable number of students still rely on financial support for their meals.
## 4. Test Preparation Course Participation:
### According to the dataset, 65% of students did not complete the test preparation course, while only 34.6% completed it. This shows that most students appeared for the test without taking any preparation course, while a smaller group chose to undergo extra preparation.
## 5. Performance by Gender:
### Female students constitute 51.3% of the average total scores, whereas male students make up 48.7%. This indicates that, on average, female students performed slightly better than male students.
## 6. Parental Marital Status:
### The majority of students come from families with married parents. While single and divorced parents are relatively common, students with widowed parents are quite rare.
## 7. Study Hours and Performance:
### The analysis shows that students who study less than 5 hours, between 5-10 hours, or more than 10 hours all tend to perform well. This suggests that the amount of study time does not have a significant impact on overall performance.
