## House Prices Prediction

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("house_prices.csv")

# Display basic info
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# -------------------------------
# GRAPH 1: Area vs Price (Scatter Plot)
# -------------------------------
plt.figure()
plt.scatter(df['Area'], df['Price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

# -------------------------------
# GRAPH 2: Average Price by Location (Bar Chart)
# -------------------------------
avg_price_location = df.groupby('Location')['Price'].mean()

plt.figure()
avg_price_location.plot(kind='bar')
plt.xlabel("Location")
plt.ylabel("Average Price")
plt.title("Average Price by Location")
plt.show()

# -------------------------------
# GRAPH 3: Price Distribution (Histogram)
# -------------------------------
plt.figure()
plt.hist(df['Price'], bins=20)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()

# -------------------------------
# EXTRA ANALYSIS (Optional but useful)
# -------------------------------

# Average price by property type
avg_price_type = df.groupby('Property_Type')['Price'].mean()
print("\nAverage Price by Property Type:")
print(avg_price_type)

# Correlation matrix
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```
